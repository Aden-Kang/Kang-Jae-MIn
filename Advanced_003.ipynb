# 05 Visualization ( 여러 개의 데이터를 하나의 그래프에 같이 그리기)

## 5-1 pie chart

```python
# 도화지 만들기
```

```python
plt.figure(figsize = (20,4))

# 파이 차트를 그리려면 각 그룹의 비율 내지는 개수, 그리고 각 그룹의 이름이 필요하다.
ratio = [len(df0), len(df1), len(df2)]
labels = ['setosa', 'versicolor', 'virginica']
# 파이 차트 그리기
plt.pie(ratio, labels = labels, autopct = '%.1f%%')
## 맨 뒤에 붙는 %%는 formatting 할 때 % 기호를 그대로 출력하라는 소리.
## 앞에 있는 %.1f는 flaating number의 소수 첫째자리까지 표시해라.
# 제목 붙이기
plt.title('Pie chart among three types of iris', fontsize = 14)
#각 색깔에 대한 정보 추가
plt.legend(bbox_to_anchor = (1,1))
#예쁘게 그리기
plt.show()
```

## 5-2 boxplot
*** df0, df1, df2에서 sepal length (cm)의 전체적인 분포(min, Q1, Q2, Q3, max)의 차이

```python
plt.figure(figsize = (6,4))
fig, ax = plt.subplots()
ax.boxplot([df0['sepal length (cm)'], df1['sepal length (cm)'], df2['sepal length (cm)']])
plt.title('Sepal length boxplot of three types of iris setosa', fontsize = 14)
plt.xticks(range(1,4), ['setosa', 'versicolor', 'virginica'])
plt.show()
```

## 5-3 histogram(겹쳐 그리기)

### 제일 간단한 버젼

```python
plt.figure(figsize = (6,4))
plt.hist(df0['sepal length (cm)'])
plt.hist(df1['sepal length (cm)'])
plt.hist(df2['sepal length (cm)'])
plt.title('Sepal length historgam of iris setosa', fontsize = 14)
plt.show()
```

### 조금 더 꾸민 버젼

```python
plt.figure(figsize = (6,4))
plt.hist(df0['sepal length (cm)'], label = 'setosa')
plt.hist(df1['sepal length (cm)'], label = 'versicolor')
plt.hist(df2['sepal length (cm)'], label = 'virginica')
plt.title('Sepal length historgam of iris setosa', fontsize = 14)
plt.legend(bbox_to_anchor = (1.35,1))
plt.show()
```

```python
plt.figure(figsize = (6,4))
plt.hist(df0['sepal length (cm)'], alpha = 0.5, label = 'setosa')
plt.hist(df1['sepal length (cm)'], alpha = 0.5, label = 'versicolor')
plt.hist(df2['sepal length (cm)'], alpha = 0.5, label = 'virginica')
plt.title('Sepal length historgam of iris setosa', fontsize = 14)
plt.legend(bbox_to_anchor = (1.35,1))
plt.show()
```
